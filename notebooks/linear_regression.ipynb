```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.rand(100, 1)
y = 3 * X + 4 + np.random.randn(100, 1) * 0.1

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data")
plt.show()
```

```python
w = np.random.randn(1)
b = np.random.randn(1)

lr = 0.1
epochs = 100
losses = []

for epoch in range(epochs):
    y_pred = w * X + b
    error = y - y_pred
    loss = (error ** 2).mean()
    losses.append(loss)

    grad_w = -2 * (X * error).mean()
    grad_b = -2 * error.mean()

    w -= lr * grad_w
    b -= lr * grad_b

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Loss over Epochs")
plt.show()
```

```python
import torch
from torch import nn
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)
```

```python
model = nn.Linear(1, 1)
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
losses = []
epochs = 100

for epoch in range(epochs):
    y_pred = model(X_tensor)
    loss = loss_fn(y_pred, y_tensor)
    losses.append(loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.4f}")
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("PyTorch Loss over Epochs")
plt.show()
```
