```python
import pandas as pd
import matplotlib.pyplot as plt

# Show all columns in outputs
pd.set_option("display.max_columns", None)
```

```python
# Test Data
test = pd.read_csv("../data/test.csv")

print("\n--- Test Data ---")
print("Shape:", test.shape)
print(test.head())
print(test.info())
print(test.describe())

print("\nMissing values in test:")
print(test.isnull().sum())

if "Basic_Demos-Age" in test.columns:
    test["Basic_Demos-Age"].hist(bins=20)
    plt.title("Age Distribution - Test")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

if "Basic_Demos-Enroll_Season" in test.columns:
    test["Basic_Demos-Enroll_Season"].value_counts().plot(kind="bar")
    plt.title("Enrollment Season Distribution - Test")
    plt.xlabel("Season")
    plt.ylabel("Count")
    plt.show()
```
